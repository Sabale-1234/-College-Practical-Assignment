```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r'C:\dsbda\Boston.csv')
print(df)
```

```python
df.info()
df.describe()
```

```python
lr=LinearRegression()
x=df.iloc[:,0].values.reshape(-1,1)
y=df.iloc[:,1].values.reshape(-1,1)
lr.fit(x,y)
y_predict=lr.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_predict,color='r')
plt.xlabel("age")
plt.ylabel("medv")
plt.title("age,medv Prediction Model")
plt.plot()
plt.show()
```
